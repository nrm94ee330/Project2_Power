```python
main_df=pd.DataFrame()

good_columns=["Year","Value"]

for st in States:
    
    
    queryA = "ELEC.GEN.ALL-" + st[1] + "-99.A"
    queryA_url = url + "api_key=" + api_key + "&series_id=" + queryA
    #print(queryA_url)    
    articlesA = requests.get(queryA_url).json()
    d=articlesA["series"][0]["data"]
    datA=pd.DataFrame(data=d,columns=good_columns)
    datA["State_Abrv"]=st[1]
    datA.to_sql(name='tbl_powerdata_annual_temp', con=engine, if_exists='replace', index=False)  
    engine.execute("INSERT IGNORE INTO tbl_powerdata_annual SELECT * FROM tbl_powerdata_annual_temp")
    print("Done Loading Annual data for " + st[1])
    
    queryM = "ELEC.GEN.ALL-" + st[1] + "-99.M"
    queryM_url = url + "api_key=" + api_key + "&series_id=" + queryM
    #print(queryM_url)
    articlesM = requests.get(queryM_url).json()
    d=articlesM["series"][0]["data"]
    datM=pd.DataFrame(data=d,columns=good_columns)
    datM["State_Abrv"]=st[1]
    datM.to_sql(name='tbl_powerdata_monthly_temp', con=engine, if_exists='replace', index=False)
    engine.execute("INSERT IGNORE INTO tbl_powerdata_monthly SELECT left(Year,4) As Year, right(Year,2) as month, Value,State_Abrv FROM tbl_powerdata_monthly_temp")
    print("Done Loading Monthly data for " + st[1])


```
